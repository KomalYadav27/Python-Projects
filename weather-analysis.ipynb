```python
#finding all records where weather is clear
df[df['Weather']=='Clear']
```

```python
#renaming a column Weather to Weather_condition
df.rename(columns = {'Weather':'Weather_condition'}, inplace='true')
```

```python
#finding records where wind speed > 24 and visibility = 25
df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)]
```

```python
df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)].shape
```

```python
#finding mean of each column against each weather condition
df.groupby('Weather_condition').mean()
```

```python
#finding records weather is clear and relative humidity is greater than 50 or visibility is above 40
df[(df['Weather_condition'] == 'Clear') & ((df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40))]
```

```python
#Weather conditions where snow is there
df.loc[df['Weather_condition'].str.contains("Snow",case = False)]
```
