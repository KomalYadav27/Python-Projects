# Gas Price Data Analysis

```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv(r'C:\Users\Klaus\Downloads\gas_prices.csv')
print(gas)
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas Prices (in USD)')
plt.plot(gas.Year,gas.USA,'b.-',label = 'USA',)
plt.plot(gas.Year,gas.Canada,'r.-',label = 'CA')
plt.plot(gas.Year,gas['South Korea'],'g.-',label = 'SK')
plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Price(in USD)')
plt.legend()
print(plt)
```
